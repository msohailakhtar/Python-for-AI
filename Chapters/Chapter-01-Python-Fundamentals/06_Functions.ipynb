# Introduction

As programs become larger, writing all the code in one place makes them difficult to understand and maintain.

Functions allow us to divide a program into smaller, reusable blocks of code.

Instead of writing the same code multiple times, we write it once inside a function and call it whenever needed.

Functions improve:

- Readability
- Reusability
- Modularity
- Maintainability
- Testing

# What is a Function?

A function is a reusable block of code that performs a specific task.

Think of a function as a machine:

Input → Processing → Output

A function can:

- Accept inputs (parameters)
- Process data
- Return a result

````python
def greet():
    print("Welcome to Python for AI!")

greet()
````

# Defining a Function

Python uses the `def` keyword to define a function.

## Syntax

```python
def function_name():
    statements
```

````python
def welcome():

    print("Hello Students!")
    print("Welcome to the Python for AI Course.")

welcome()
````

# Function Parameters

Parameters allow information to be passed into a function.

Example:

```python
def greet(name):
    print(name)
```

````python
def greet(name):
    print(f"Hello, {name}!")

greet("Sohail")
greet("Ali")
greet("Ayesha")
````

# Function with Return Value

The `return` statement sends a value back to the caller.

````python
def square(number):

    return number ** 2

result = square(8)

print(result)
````

# Multiple Parameters

Functions can receive multiple parameters.

````python
def add(a, b):

    return a + b

print(add(10, 20))
print(add(50, 60))
````

# Default Parameters

A default parameter has a predefined value.

````python
def greet(name="Student"):

    print(f"Welcome {name}")

greet()
greet("Sohail")
````

# Local and Global Variables

Variables defined inside a function are local.

Variables defined outside are global.

````python
course = "Python for AI"

def display():

    student = "Sohail"

    print(course)
    print(student)

display()
````

# Lambda Functions

Lambda functions are small anonymous functions.

Syntax:

```python
lambda arguments: expression
```

````python
square = lambda x: x ** 2

print(square(10))
````

# Recursive Functions

A recursive function calls itself until a stopping condition is reached.

````python
def factorial(n):

    if n == 1:
        return 1

    return n * factorial(n - 1)

print(factorial(5))
````

# AI Insight

Functions are fundamental in AI.

Examples:

- Data preprocessing
- Model training
- Prediction
- Evaluation
- Feature engineering
- Neural network layers

Every major AI library is built around reusable functions.

````python
def train_model(epochs):

    for epoch in range(1, epochs + 1):
        print(f"Epoch {epoch} completed")

train_model(5)
````

# Best Practices

- Keep functions short.
- Give meaningful names.
- Write one task per function.
- Use return instead of print when appropriate.
- Add comments and documentation.
