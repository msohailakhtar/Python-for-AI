# Introduction

Conditional statements allow a program to make decisions.

Instead of executing every statement, Python checks a condition first.

If the condition is **True**, one block of code executes.

Otherwise, another block executes.

Decision making is one of the most important concepts in programming because it enables programs to respond differently depending on the input or situation.

# Flow of a Conditional Statement

```
Condition
    │
    ▼
 True? ─────► Execute Block
    │
    ▼
 False
    │
    ▼
Execute Alternative Block
```

Conditional statements help programs choose different paths based on conditions.

# The if Statement

The **if** statement executes a block of code only when the condition is **True**.

### Syntax

```python
if condition:
    statement
```

````python
age = 20

if age >= 18:
    print("You are eligible to vote.")
````

# The if...else Statement

The **else** block executes when the condition is **False**.

### Syntax

```python
if condition:
    statement
else:
    statement
```

````python
marks = 45

if marks >= 50:
    print("Pass")
else:
    print("Fail")
````

# The if...elif...else Statement

Use **elif** when checking multiple conditions.

Python evaluates the conditions from top to bottom and executes the first condition that is True.

````python
marks = 82

if marks >= 90:
    print("Grade A+")

elif marks >= 80:
    print("Grade A")

elif marks >= 70:
    print("Grade B")

elif marks >= 60:
    print("Grade C")

else:
    print("Fail")
````

# Nested if Statements

An **if statement inside another if statement** is called a nested if statement.

Nested conditions are useful when one decision depends on another.

````python
age = 22
cgpa = 3.60

if age >= 18:
    if cgpa >= 3.0:
        print("Eligible for Scholarship")
    else:
        print("CGPA is too low")
else:
    print("Not eligible")
````

# Using Logical Operators

Logical operators combine multiple conditions.

| Operator | Meaning |
|----------|---------|
| **and** | Both conditions must be True |
| **or** | At least one condition must be True |
| **not** | Reverses the condition |

````python
temperature = 28
is_raining = False

if temperature < 30 and not is_raining:
    print("Good weather for a walk.")
else:
    print("Better stay indoors.")
````

# Mathematics Behind Conditional Statements

Conditional statements are based on **Boolean Logic**.

Every condition evaluates to either:

- **True**
- **False**

Examples:

- 10 > 5 → True
- 5 == 7 → False

Boolean logic is fundamental to mathematics, digital circuits, and Artificial Intelligence.

````python
print(10 > 5)
print(7 == 9)
print(15 != 8)
````

# Conditional Statements in Artificial Intelligence

AI systems constantly make decisions.

Examples include:

- Spam Detection
- Face Recognition
- Loan Approval
- Disease Diagnosis
- Autonomous Vehicles
- Recommendation Systems

These systems evaluate conditions before taking actions.

````python
prediction = 0.91

if prediction >= 0.80:
    print("Positive Prediction")
else:
    print("Negative Prediction")
````
