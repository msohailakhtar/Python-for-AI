# Input in Python

The `input()` function allows a program to receive information from the user.

Syntax:

```python
name = input("Enter your name: ")
```

The value returned by `input()` is always a string unless converted into another data type.

````python
name = input("Enter your name: ")

print("Welcome,", name)
````

# Output in Python

Python uses the `print()` function to display information.

Syntax:

```python
print("Hello World")
```

The print() function can display:

- Strings
- Numbers
- Variables
- Mathematical expressions
- Multiple values

````python
name = "Sohail"
age = 32

print(name)
print(age)

print("Name:", name)
print("Age:", age)

print("Welcome to Python for AI")
````

# Formatted Output

Python provides formatted output using:

- Comma separation
- f-strings
- format() method

Among these, **f-strings** are the most recommended because they are readable and efficient.

````python
name = "Ali"
age = 24
cgpa = 3.78

print(f"Student Name : {name}")
print(f"Age          : {age}")
print(f"CGPA         : {cgpa}")
````

# Type Casting

Type casting converts one data type into another.

Python supports two types of type casting:

1. Implicit Type Casting
2. Explicit Type Casting

# Explicit Type Casting

The programmer manually converts data types.

Common functions:

| Function | Converts To |
|----------|-------------|
| int( ) | Integer |
| float( ) | Floating Point |
| str( ) | String |
| bool( ) | Boolean |

````python
age = input("Enter your age: ")

print(type(age))

age = int(age)

print(type(age))

print(age + 5)
````

# Implicit Type Casting

Python automatically converts compatible data types during calculations.

Example:

An integer can automatically become a float during arithmetic operations.

````python
a = 10
b = 2.5

result = a + b

print(result)

print(type(result))
````

# Common Type Conversion Functions

| Function | Example | Result |
|----------|---------|--------|
| int( ) | int(4.8) | 4 |
| float( ) | float(5) | 5.0 |
| str( ) | str(100) | "100" |
| bool( ) | bool(1) | True |
| list( ) | list("AI") | ['A','I'] |

````python
print(int(4.8))

print(float(5))

print(str(100))

print(bool(1))

print(bool(0))
````

# Type Casting in Artificial Intelligence

AI systems continuously convert data types.

Examples include:

- Reading numerical datasets
- Loading CSV files
- Processing images
- Neural network calculations
- Feature engineering

Correct type conversion prevents runtime errors and improves model performance.

````python
learning_rate = float(input("Learning Rate: "))

epochs = int(input("Epochs: "))

accuracy = float(input("Accuracy: "))

print("\nTraining Configuration")

print("------------------------")

print(f"Learning Rate : {learning_rate}")

print(f"Epochs        : {epochs}")

print(f"Accuracy      : {accuracy}")
````

# Best Practices

- Always validate user input.
- Convert input into the required data type.
- Use descriptive prompts.
- Prefer f-strings for output formatting.
- Handle invalid input using exception handling (covered later).

# Common Mistakes

Avoid these mistakes:

- Forgetting that input() returns a string.
- Performing arithmetic without conversion.
- Mixing incompatible data types.
- Ignoring invalid user input.
