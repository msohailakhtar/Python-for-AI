# Introduction

A loop is a programming structure that allows a block of code to execute repeatedly until a specified condition is met.

Instead of writing the same code many times, a loop automates repetitive tasks.

Loops make programs:

- Shorter
- Faster
- Easier to maintain
- More efficient

Loops are essential in almost every software application, from simple calculators to advanced Artificial Intelligence systems.

# Why Do We Need Loops?

Imagine printing:

Welcome to Python for AI

100 times.

Without loops, you would need to write 100 print() statements.

With loops, you only need two lines of code.

Loops save time, reduce errors, and make programs scalable.

````python
for i in range(5):
    print("Welcome to Python for AI")
````

# The for Loop

The **for loop** is used to iterate over a sequence such as:

- Numbers
- Strings
- Lists
- Tuples
- Dictionaries
- Sets

## Syntax

```python
for variable in sequence:
    statements
```

````python
for number in range(1, 6):
    print(number)
````

# The range() Function

The `range()` function generates a sequence of numbers.

Examples:

```python
range(5)

range(1,6)

range(2,21,2)
```

Arguments:

- Start
- Stop
- Step

````python
print("Numbers from 1 to 10")

for i in range(1, 11):
    print(i)
````

# Printing Even Numbers

Use a step size of 2 to print only even numbers.

````python
for i in range(2, 21, 2):
    print(i)
````

# The while Loop

A while loop executes as long as its condition remains True.

Syntax:

```python
while condition:
    statements
```

````python
count = 1

while count <= 5:
    print(count)
    count += 1
````

# Infinite Loops

An infinite loop never stops because its condition never becomes False.

Example:

```python
while True:
    print("Infinite Loop")
```

Be careful when writing while loops.

# The break Statement

The `break` statement immediately terminates a loop.

````python
for i in range(1, 11):

    if i == 6:
        break

    print(i)
````

# The continue Statement

The `continue` statement skips the current iteration and moves to the next one.

````python
for i in range(1, 11):

    if i == 5:
        continue

    print(i)
````

# Nested Loops

A loop inside another loop is called a nested loop.

Nested loops are useful for:

- Matrix operations
- Pattern printing
- Tables
- Games

````python
for row in range(3):
    for column in range(3):
        print("*", end=" ")
    print()
````

# Mathematical Example

Loops are widely used in mathematics.

Examples:

- Sum of numbers
- Factorial
- Fibonacci sequence
- Prime numbers

````python
total = 0

for i in range(1, 101):
    total += i

print("Sum =", total)
````

# AI Insight

Loops are everywhere in Artificial Intelligence.

Examples:

- Reading millions of images.
- Training neural networks for many epochs.
- Processing datasets.
- Updating model parameters.
- Running reinforcement learning episodes.

Without loops, modern AI systems would not be possible.

````python
epochs = 5

for epoch in range(1, epochs + 1):
    print(f"Training Epoch {epoch} Completed")
````

# Best Practices

- Keep loops simple.
- Avoid unnecessary nested loops.
- Use meaningful variable names.
- Avoid infinite loops.
- Use break and continue only when necessary.
