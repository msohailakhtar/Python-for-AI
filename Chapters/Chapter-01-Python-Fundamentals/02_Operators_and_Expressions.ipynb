# Introduction

Operators are one of the most fundamental components of any programming language.

An **operator** is a symbol that tells Python to perform a specific operation on one or more values (called operands).

For example:

- Adding two numbers
- Comparing two values
- Assigning a value to a variable
- Checking logical conditions

Expressions are combinations of variables, values, and operators that Python evaluates to produce a result.

Understanding operators is essential because they are used in almost every Python program, from simple calculations to advanced Artificial Intelligence algorithms.

# What is an Operator?

# Arithmetic Operators

Arithmetic operators perform mathematical calculations.

| Operator | Description | Example |
|-----------|-------------|---------|
| **+** | Addition | `10 + 5 = 15` |
| **-** | Subtraction | `10 - 5 = 5` |
| **\*** | Multiplication | `10 * 5 = 50` |
| **/** | Division | `10 / 5 = 2.0` |
| **//** | Floor Division | `10 // 3 = 3` |
| **%** | Modulus (Remainder) | `10 % 3 = 1` |
| **\*\*** | Exponent | `2 ** 3 = 8` |

````python
# Arithmetic Operators

a = 20
b = 6

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** 2)
````

# Comparison Operators

Comparison operators compare two values and return either **True** or **False**.

| Operator | Description | Example |
|-----------|-------------|---------|
| **==** | Equal to | `10 == 10` |
| **!=** | Not Equal to | `10 != 5` |
| **>** | Greater Than | `10 > 5` |
| **<** | Less Than | `5 < 10` |
| **>=** | Greater Than or Equal To | `10 >= 5` |
| **<=** | Less Than or Equal To | `5 <= 10` |

````python
x = 15
y = 10

print("Equal:", x == y)
print("Not Equal:", x != y)
print("Greater:", x > y)
print("Less:", x < y)
print("Greater or Equal:", x >= y)
print("Less or Equal:", x <= y)
````

# Assignment Operators

Assignment operators assign or update the value of a variable.

| Operator | Example | Equivalent |
|-----------|---------|------------|
| **=** | `x = 5` | Assign value |
| **+=** | `x += 3` | `x = x + 3` |
| **-=** | `x -= 3` | `x = x - 3` |
| ***=** | `x *= 3` | `x = x * 3` |
| **/=** | `x /= 3` | `x = x / 3` |
| **%=** | `x %= 3` | `x = x % 3` |
| **\*\*=** | `x **= 3` | `x = x ** 3` |

````python
x = 10

print("Initial Value:", x)

x += 5
print("After += :", x)

x -= 2
print("After -= :", x)

x *= 3
print("After *= :", x)

x /= 2
print("After /= :", x)

x %= 4
print("After %= :", x)
````

# Logical Operators

Logical operators are used to combine multiple conditions.

They always return either **True** or **False**.

| Operator | Description | Example |
|-----------|-------------|---------|
| **and** | Returns **True** if both conditions are True | `x > 5 and y > 5` |
| **or** | Returns **True** if at least one condition is True | `x > 5 or y > 5` |
| **not** | Reverses the Boolean value | `not(x > 5)` |

---

## Mathematics Behind Logical Operators

Logical operators are based on **Boolean Algebra**, developed by George Boole.

They are fundamental in:

- Artificial Intelligence
- Machine Learning
- Decision Making
- Digital Logic

````python
age = 22
cgpa = 3.65

print("AND:", age > 18 and cgpa > 3.0)

print("OR :", age > 25 or cgpa > 3.0)

print("NOT:", not(age > 30))
````

# Identity Operators

Identity operators compare whether two variables refer to the **same object in memory**.

| Operator | Description |
|-----------|-------------|
| **is** | Returns True if both variables refer to the same object |
| **is not** | Returns True if they refer to different objects |


> **Note:** Use **==** to compare values and **is** to compare object identity.

````python
a = [1, 2, 3]

b = a

c = [1, 2, 3]

print("a is b :", a is b)

print("a is c :", a is c)

print("a == c :", a == c)
````

# Membership Operators

Membership operators check whether a value exists inside a sequence.

| Operator | Description |
|-----------|-------------|
| **in** | Returns True if the value exists |
| **not in** | Returns True if the value does not exist |

Membership operators work with:

- Lists
- Tuples
- Strings
- Dictionaries
- Sets

````python
fruits = ["Apple", "Banana", "Orange"]

print("Apple" in fruits)

print("Mango" in fruits)

print("Banana" not in fruits)
````

# Bitwise Operators

Bitwise operators perform operations directly on the binary representation of integers.

| Operator | Description |
|-----------|-------------|
| **&** | Bitwise AND |
|       | Bitwise OR |
| **^** | Bitwise XOR |
| **~** | Bitwise NOT |
| **<<** | Left Shift |
| **>>** | Right Shift |

Although beginners rarely use bitwise operators, they are important in low-level programming, embedded systems, and optimization.

````python
a = 5

b = 3

print("AND :", a & b)

print("OR :", a | b)

print("XOR :", a ^ b)

print("NOT :", ~a)

print("Left Shift :", a << 1)

print("Right Shift :", a >> 1)
````

# Operator Precedence

Operator precedence determines the order in which operations are evaluated.

Higher precedence operators are evaluated before lower precedence operators.

| Priority | Operator |
|----------|----------|
| 1 | **()** Parentheses |
| 2 | **** Exponent |
| 3 | ***, /, //, %** |
| 4 | **+, -** |
| 5 | Comparison Operators |
| 6 | **not** |
| 7 | **and** |
| 8 | **or** |

Using parentheses improves readability and avoids mistakes.

````python
print(10 + 5 * 2)

print((10 + 5) * 2)

print(2 ** 3 * 4)

print((2 ** 3) * 4)
````

Examples of AI variables:
