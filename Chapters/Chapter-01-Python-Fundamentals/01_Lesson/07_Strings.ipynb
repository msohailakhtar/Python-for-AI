# Introduction

Strings are one of the most commonly used data types in Python.

A string is a sequence of characters enclosed in single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """).

Strings are immutable, which means their characters cannot be changed after creation.

They are essential for processing text, handling user input, reading files, and developing AI applications.

# Creating Strings

Python provides several ways to create strings.

Examples:

- Single quotes
- Double quotes
- Triple quotes (multi-line strings)

````python
name = "Muhammad Sohail Akhtar"

course = 'Python for AI'

description = """
Artificial Intelligence
Machine Learning
Data Science
"""

print(name)
print(course)
print(description)
````

# String Indexing

Each character in a string has an index.

Python starts indexing from **0**.

Example:

P  y  t  h  o  n

0  1  2  3  4  5

````python
language = "Python"

print(language[0])
print(language[3])
print(language[5])
````

# Negative Indexing

Negative indexing starts from the end of the string.

P  y  t  h  o  n

-6 -5 -4 -3 -2 -1

````python
language = "Python"

print(language[-1])
print(language[-2])
print(language[-6])
````

# String Slicing

Slicing extracts a portion of a string.

Syntax:

```python
string[start:stop:step]
```

````python
course = "Python for Artificial Intelligence"

print(course[0:6])
print(course[7:10])
print(course[:6])
print(course[-12:])
````

# String Methods

Python provides many built-in methods to manipulate strings.

Common methods include:

- upper( )
- lower( )
- title( )
- capitalize( )
- strip( )
- replace( )
- split( )
- join( )
- find( )
- count( )

````python
text = "python for ai"

print(text.upper())
print(text.title())
print(text.capitalize())
````

# Searching and Replacing

Use `find()` to locate text and `replace()` to modify it.

````python
sentence = "Artificial Intelligence is amazing."

print(sentence.find("Intelligence"))

new_sentence = sentence.replace("amazing", "powerful")

print(new_sentence)
````

# String Formatting

Formatted strings (f-strings) provide a clean and readable way to insert variables into text.

````python
name = "Sohail"
cgpa = 3.80

print(f"Student: {name}")
print(f"CGPA: {cgpa}")
````
